# Data Visualisation

```python
sns.countplot(x='fixed acidity',data=df)
plt.show()
```

```python
### Let us divide our data into x and y where x are independent variables and y are dependent variables.
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# Now we will split our dataset into train and test data set
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# Now we have successfully divided our dataset into training and testing dataset
```

```python
# Now we have to model the logistic regression
# All the model follows the same 4 step or follows the same 4 pattern
# 1.Call the model
# 2.Create an object for the model
# 3. Fit the model 
# 4. Then final predict the model
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
# Now we are going to calculate the accuracy score
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```
